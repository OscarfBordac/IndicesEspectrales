```python
import rasterio
from rasterio import plot
import numpy as np
```

```python
#Calculo NDVI

ruta_entrada = r"E:\Trabajo\Tesis\mosaicos\202212\MosaicoL9_202212_202301.tif"
ruta_salida = r"E:\Trabajo\Tesis\SR\Indices\MosaicoL9_202212_202301_ndvi.tif"

with rasterio.open(ruta_entrada) as src:
    red_band = src.read(4).astype(np.float32)  # Band 3: Red
    nir_band = src.read(5).astype(np.float32)  # Band 4: Near Infrared
    profile = src.profile

mask = np.logical_or(red_band == 0, nir_band == 0)
red_band[mask] = 0
nir_band[mask] = 0

ndvi = np.divide(nir_band - red_band, nir_band + red_band, where=(nir_band + red_band) != 0)
profile.update(dtype=rasterio.float32, count=1)
with rasterio.open(ruta_salida, 'w', **profile) as dst:
    dst.write(ndvi, 1)
```

```python
#NBR
ruta_entrada = r"E:\Trabajo\Tesis\mosaicos\202212\MosaicoL9_202212_202301.tif"
ruta_salida = r"E:\Trabajo\Tesis\SR\Indices\NBR\MosaicoL9_202212_202301_NBR.tif"

with rasterio.open(ruta_entrada) as src:
    nir_band = src.read(5).astype(np.float32)  # Band 5: Red
    swir_band = src.read(7).astype(np.float32)  # Band 7: Near Infrared
    profile = src.profile

mask = np.logical_or(nir_band == 0, swir_band == 0)
nir_band[mask] = 0
swir_band[mask] = 0

nbr = np.divide(nir_band - swir_band, nir_band + swir_band, where=(nir_band + swir_band) != 0)
profile.update(dtype=rasterio.float32, count=1)
with rasterio.open(ruta_salida, 'w', **profile) as dst:
    dst.write(nbr, 1)
```

```python
#msavi
ruta_entrada = r"E:\Trabajo\Tesis\SR\2021\PSH\tif\v2\LC08_L1TP_202102_T1_SR_PSH.tif"
ruta_salida = r"E:\Trabajo\Tesis\SR\Indices\MSAVI\LC08_L1TP_202102_T1_SR_PSH_msavi.tif"

with rasterio.open(ruta_entrada) as src:
    red_band = src.read(4).astype(np.float32)  # Band 4: Red
    nir_band = src.read(5).astype(np.float32)  # Band 5: Near Infrared
    profile = src.profile

# Calcular el MSAVI
msavi_numerator = 2 * nir_band + 1 - np.sqrt((2 * nir_band + 1)**2 - 8 * (nir_band - red_band))
msavi_denominator = 2

msavi = np.divide(msavi_numerator, msavi_denominator, where=msavi_denominator != 0)

# Asegurarse de que los valores estén dentro del rango válido para el MSAVI (-1 a 1)
msavi = np.clip(msavi, -1, 1)

# Guardar el resultado como una nueva imagen
with rasterio.open(ruta_entrada) as src:
    profile = src.profile
    profile.update(dtype=rasterio.float32, count=1)

    with rasterio.open(ruta_salida, 'w', **profile) as dst:
        dst.write(msavi, 1)
```

```python
#EVI
ruta_entrada = r"E:\Trabajo\Tesis\mosaicos\202212\MosaicoL9_202212_202301.tif"
ruta_salida = r"E:\Trabajo\Tesis\SR\Indices\EVI\MosaicoL9_202212_202301_EVI.tif"

# Coeficientes del EVI
C1 = 6
C2 = 7.5
L = 1

with rasterio.open(ruta_entrada) as src:
    red_band = src.read(4).astype(np.float32)  # Band 4: Red
    nir_band = src.read(5).astype(np.float32)  # Band 5: Near Infrared
    blue_band = src.read(2).astype(np.float32)  # Band 2: Blue

# Aplicar la fórmula del EVI
numerator = nir_band - red_band
denominator = nir_band + C1 * red_band - C2 * blue_band + L

# Para evitar divisiones por cero, donde el denominador sea cero, el resultado será cero.
EVI = np.divide(numerator, denominator, where=(denominator != 0), out=np.zeros_like(numerator))

# Asegurarse de que los valores estén dentro del rango válido para el EVI (-1 a 1)
EVI = np.clip(EVI, -1, 1)

# Guardar el resultado como una nueva imagen
with rasterio.open(ruta_entrada) as src:
    profile = src.profile
    profile.update(dtype=rasterio.float32, count=1)

    with rasterio.open(ruta_salida, 'w', **profile) as dst:
        dst.write(EVI, 1)
```
